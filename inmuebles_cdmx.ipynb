### **Creando Filtros**

#### 1. Apartamentos que tienen 1 dormitorio y un alquiler menor a MXN 4200.

```python
seleccion1= df['Habitaciones'] ==1
#Aplicamos el filtro que habitaciones sean de 1
```

```python
type(seleccion1)
```

```python
df[seleccion1]
#aquí visualizamos las que tienen solo una habitación
```

```python
seleccion2= df['Valor'] <4200
df[seleccion2]
#Aquí aplicamos otro filtro que el valor sea menor a 4200
```

```python
df[seleccion2].max()
#aquí verificamos que el valor máximo sea menor a 4200
```

```python
filtro1 = (seleccion1) & (seleccion2)
df[filtro1]
#Aquí aplicamos los dos filtros anteriores
```

```python
df1= df[filtro1]
```

#### 2. Apartamentos que tienen al menos 2 dormitorios, un alquiler menor a MXN 10500 y una superficie mayor a 70 m².

```python
filtro2=(df['Habitaciones'] >=2) & (df['Valor'] < 10500) & (df['Area'] > 70)
df[filtro2]
#Aqui aplicamos en un solo filtro las 3 cosas que solicitan 2 habitaciones menores a 10500 con una sup 70 m2
```

```python
df2= df[filtro2]
```
